```python

import numpy as np
import random
```

```python
def cor(links):
  ones = np.ones(len(links))
  sources = [x[0] for x in links]
  targets = [x[1] for x in links]
  n = max(max(sources), max(targets)) + 1
  return(sources,targets)
```

```python
#transition matrix of sources and targets (according to given links)
def get_mat(sources,targets,tot_nodes):
  #initilize matrix with all 0's in starting
  matr = [ [ 0 for i in range(tot_nodes) ] for j in range(tot_nodes) ]
  li_stp =[]#list of Source Target and probability
  for i in range(len(sources)): #length of sources and targets are equal
      s_cor = sources[i]
      t_cor = targets[i]
      #print(s_cor,t_cor)
      prob =round(1/sources.count(s_cor),5)
      matr[t_cor][s_cor] =prob
      str =s_cor,t_cor,prob
      li_stp.append(str)
  #if a 
  for n in range(tot_nodes):
    if(sources.count(n) == 0):
      for i in range(tot_nodes):
        matr[i][n] = 1/tot_nodes
  return(matr,li_stp)    
```

```python
def get_next_vec(vec,tot_nodes,mat):

  n_vec = vec
  for i in range(tot_nodes):
    mul_sum=0
    
    for j in zip(mat[i],vec):
      mul_sum += j[0]*j[1]

    n_vec[i]=round(mul_sum,4)

  return(n_vec)
```

```python
#generate random tuples
links_li =[]
for i in range(0,random.randint(1000,10000)):
  srce = random.randint(0,1000)
  tar = random.randint(0,1000)
  if srce != tar:
    tup =(srce,tar)
    links_li.append(tup)
```
